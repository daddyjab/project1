```python
#Import Pandas and read csv file
import pandas as pd
import numpy as np
df = pd.read_csv("Business_Licenses.csv")
df.head()
```

```python
df.shape
```

```python
# List all variables as some were omited above. 
list(df)
```

```python
# Create new dataframe with the variables we'll use. 
df_1 = df[["LEGAL NAME", 'DOING BUSINESS AS NAME', "LICENSE CODE", "CITY", "WARD", "ZIP CODE", "DATE ISSUED", "LATITUDE", "LONGITUDE"]]
df_1.head()
```

```python
# Flter by city (Chicago) 
df_2 = df_1.query('CITY == "CHICAGO"')
df_2.head()

```

```python
#Remaining rows
df_2.shape
```

```python
# Filter by license code
df_3 = df_2[df_2['LICENSE CODE'] == 1006]

#Reamining rows
df_3.shape
```

```python
#Print head 
df_3.head(50)
```

```python
# Drop missing data
df_4 = df_3.dropna()

#Remaining rows
df_4.shape
```

```python
# Print head
df_4 = df_4.reset_index(drop=True)
df_4.head()
```

```python
# Save clean dataset into a CSV file. 
df_4.to_csv('Clean_Chicago_Restaurants_Licenses.csv')
```
